```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
earthquake = pd.read_csv('data/earthquake_data.csv')
```

```python
earthquake.head()


earthquake.time = pd.to_datetime(earthquake['time'])

#earthquake['year'] = earthquake[earthquake['time']].year
```

```python
countries = gpd.read_file('data/ne_110m_admin_0_countries')
countries.head()
```

```python
countries.plot()
```

```python
earthquake.plot(x = 'longitude',
               y = 'latitude',
               kind = "scatter",
               alpha = 0.5)
```

```python
big_earthquake = earthquake[(earthquake['mag'] >= 6.5)]

big_earthquake.head()
big_earthquake.plot(x = 'longitude',
               y = 'latitude',
               kind = "scatter",
               alpha = 0.5)
```

```python
# final figure
fig, ax = plt.subplots()
countries.plot(ax = ax,
              color = "lightgrey")
big_earthquake.plot(ax=ax,
                   column = 'mag',
                   x = 'longitude',
                   y = 'latitude',
                   kind = "scatter",
                   alpha = 0.5, 
                   edgecolor = "red")
fig.text(0.1, 0.1, 'Only Author: STEPHAN')
```
